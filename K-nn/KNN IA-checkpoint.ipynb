### Classificando com KNN

Importando KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

Criando um identificador
Pelo numero de vizinhos 
que ele deve levar em consideração
(0,1,2)


```python
knn = KNeighborsClassifier(n_neighbors=3)
```

Treinando Identificador 

```python
knn.fit(X,y)
```

Carregando arquivo Teste

```python
ndf = pd.read_csv('teste.csv')
```

```python
ndf
```

```python
X1 = np.array(ndf)
X1
```

## Resultado Knn


```python
knn.predict(X1)
```
